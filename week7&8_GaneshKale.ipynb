## Filter out missing data

```python
# check number of missing data in each column

df.isnull().sum()

# there are columns with all values are null or na
```

```python
# filter out missing data from dataset

# keep columns with 70% not-na values

col_thres = round(0.7*len(df))

not_na_df = df.dropna(thresh=col_thres, axis=1)   # drop columns with not-na values less than 4005 and store in new df
```

```python
# display new dataset shape and head

not_na_df.shape
not_na_df.head()

# there were 33 columns with not-na values less than 4005
```

```python
# check number of missing data in each column on new dataset

not_na_df.isnull().sum()
```

```python
# to filter out all missing data from dataset, this can be done if features with na values are not required for analysis

clean_df = not_na_df.dropna()     #remove all rows with na values
```

```python
clean_df.shape
clean_df.head()
```

```python
# check number of missing data in each column on new dataset

clean_df.isnull().sum()
```

## Fill in Missing Data

```python
# Lets say we dropped columns which has less than 70% not-na values

#  for the columns with na values present can be filled in with appropriate values based on feature

# fill na values in data frame - not_na_df

# head of df
not_na_df.head()
```

```python
# display columns with na values only

temp = not_na_df.isna().sum()[not_na_df.isna().sum()>0]
temp
```

```python
# display columns with na values from the dataset

na_cols = list(temp.index)
not_na_df[na_cols].head()
```

```python
# check the distribution of values in each column

for col in na_cols:
    print(f'Distribution of: {col}')
    not_na_df[col].value_counts()
    print('*'*40)
```

```python
# fill in missing data in each column based on above distribution, value with most count

fill_na_dict = {'Repository':not_na_df['Repository'].value_counts().index[0],
                'Country':not_na_df['Country'].value_counts().index[0],
               'Geography Type':not_na_df['Geography Type'].value_counts().index[0],
               'Credit Line':not_na_df['Credit Line'].value_counts().index[0],
               'Dimensions':not_na_df['Dimensions'].value_counts().index[0],
               'Object Date':not_na_df['Object Date'].value_counts().index[0],
               'Culture':not_na_df['Culture'].value_counts().index[0],
               'AccessionYear':not_na_df['AccessionYear'].value_counts().index[0],
               'Gallery Number':not_na_df['Gallery Number'].value_counts().index[0]}
```

```python
# fill na values based on above dictionary - filling na with top value by count from that column

fill_na_df = not_na_df.fillna(value=fill_na_dict)
```

```python
# display head of dataset after filling na values

fill_na_df.head()
```

```python
# display na values if any after filling them

fill_na_df.isna().sum()

# no columns have na values now
```

```python
# display shape and head of dataset

multi_index = removed_dups_df.drop('index',axis=1)

multi_index.shape
multi_index.head()
```

## Create hierarchical index

```python
# create another dataset from df for multi indexing
# separate columns with int or float datatypes from categorical or bool/object

numeric_cols = ['Object Number']    # to store numeric features

cat_cols = []     # to store categorical and bool features

for col in multi_index.columns:
    if multi_index[col].dtype =='O' or multi_index[col].dtype =='bool':
        cat_cols.append(col)
    else:
        numeric_cols.append(col)
```

```python
# create two datasets

# the relationship between two dataset is column 'Object Number'

multi_index_num = multi_index[numeric_cols]

multi_index_cat = multi_index[cat_cols]
```

```python
multi_index_num.drop(['dim_inch','dim_cm'],inplace=True,axis=1)   # drop these columns as all nan values
```

```python
# display shape and head of value_countssets

multi_index_num.shape
multi_index_num.head()
```

```python
# display shape and head of two datasets

multi_index_cat.shape
multi_index_cat.head()
```

```python
# create hierarchical index of numerical df

hier_idx = multi_index_num.set_index(['Gallery Number','Object ID'],drop=True)
```

```python
# display head of multi-index df

hier_idx.head(10)
```

```python
# display tail of multi-index df

hier_idx.tail(10)
```

```python
# multi-index of categorical df

hier_idx_cat = multi_index_cat.set_index(['Culture','Country'])
```

```python
# display head of new multi-index df

hier_idx_cat.head()
```

```python
# display tail of new multi-index df

hier_idx_cat.tail()
```

## Combine and Merge Datasets

```python
# we will combine multi_index_cat and multi_index_num into one
# the relation in both columns is key 'Object Number'

comb_df = pd.merge(multi_index_num,multi_index_cat, on='Object Number')
```

```python
# display shape and head of combined df

comb_df.shape
comb_df.head()
```

## Pivot the data

```python
# create new df to for pivoting combined df

pvt_df = comb_df[['Object Begin Date','Culture','Gallery Number']]
```

```python
# update culture column to have few unique cultures

pvt_df['Culture'] =  pvt_df['Culture'].apply(lambda x : ''.join(x.split()[0]).replace(',',''))
```

```python
# display head of df

pvt_df.drop_duplicates(['Object Begin Date'],inplace=True)
pvt_df.head()
```

```python
# pivot df  - long to wide format

pivoted = pvt_df.pivot('Object Begin Date','Culture','Gallery Number')
```

```python
# display pivot df

pivoted.shape
pivoted.head()
```

### Grouping with Dicts/Series

```python
# change column from float/bool to int value

comb_df['AccessionYear'] = comb_df['AccessionYear'].astype(int)  
comb_df['Is Highlight'] = comb_df['Is Highlight'].astype(int)
comb_df['Is Public Domain'] = comb_df['Is Highlight'].astype(int)
comb_df['Is Timeline Work'] = comb_df['Is Highlight'].astype(int)
comb_df.set_index('Gallery Number',inplace=True)
# display shape and head of combined df

comb_df.shape
comb_df.head(2)
```

```python
# create dictionary with column names for grouping

map_dict = {'AccessionYear':'year','Object Begin Date':'year','Object End Date':'year'
            ,'Is Highlight':'avg','Is Public Domain':'avg','Is Timeline Work':'avg'}
```

```python
# group using mapping dict

grouped = comb_df[['AccessionYear','Object Begin Date','Object End Date','Is Highlight','Is Public Domain','Is Timeline Work']].groupby(map_dict,axis=1)
```

```python
# perform aggregation to get min and max year

grouped.min()
grouped.max()
```

### Split/Apply/Combine

```python
# function to return top 5 arts by AccessionYear 

def top_arts_5(df,col='AccessionYear'):
    top_5 = df.sort_values(by=col,ascending=False).head()
    return top_5
```

```python
# create new dataframe to view the Object Name and access year

df = comb_df[['Gallery Number','AccessionYear','Object Begin Date','Object End Date','Object Name']]
df.head()
```

```python
# view top 5 arts name each year

df.groupby('Object Name').apply(top_arts_5)
```

### Convert between string and date time

```python
# display df head

clean_df.head()
```

```python
# create data column based on accee year column

clean_df['AccessionYear_date'] = clean_df['AccessionYear'].apply(lambda x:str(int(x))+'-'+str(np.random.randint(1,13))+'-'+str(np.random.randint(1,28)))
clean_df['AccessionYear_date'] = clean_df['AccessionYear_date'].astype(str)
```

```python
# new column head and type

clean_df['AccessionYear_date'].head()
clean_df['AccessionYear_date'].dtype
```

```python
# convert string date in to datetime

from datetime import datetime

clean_df['AccessionYear_date'] = clean_df['AccessionYear_date'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d'))
```

```python
# new column head and type

clean_df['AccessionYear_date'].head(10)
clean_df['AccessionYear_date'].dtype
```

### Generate date range

```python
# create new column - data range from AccessionYear_date

clean_df['date_range'] = pd.date_range(clean_df['AccessionYear_date'].mean(),clean_df['AccessionYear_date'].max(),len(clean_df['AccessionYear_date']))
```

```python
# display the new column date_range head

clean_df['date_range'].head()
```

```python
# change the date_range column format

clean_df['date_range'] = clean_df['date_range'].apply(lambda x:x.strftime('%Y-%m-%d'))
```
